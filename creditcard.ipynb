```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

```python
credit_data = pd.read_csv("/content/drive/MyDrive/creditcard.csv")
credit_data.head()
```

```python
credit_data.info()
```

```python
credit_data.isnull().sum()
```

```python
credit_data["Class"].value_counts()
```

```python
legit =credit_data[credit_data.Class==0]
fraud = credit_data[credit_data.Class==1]
legit.shape
```

```python
legit['Amount'].describe()
```

```python
fraud['Amount'].describe()
```

```python
credit_data.groupby('Class').mean()
```

```python
legit_sample= legit.sample(492)
```

```python
new_data = pd.concat([legit_sample,fraud], axis =0)
new_data.head()
```

```python
x = new_data.drop(columns='Class',axis=1)
y = new_data['Class']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2,stratify=y,random_state=2)
```

```python
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
x_train_prediction = model.predict(x_train)
training_accuracy = accuracy_score(x_train_prediction, y_train)
print('Accuracy on Training data: ', training_accuracy)
```

```python
x_test_predict = model.predict(x_test)
testing_accuracy = accuracy_score(x_test_predict,y_test)
print('Accuracy on Testing data: ', testing_accuracy)
```

```python
from sklearn.ensemble import RandomForestClassifier  # Import the necessary class
model1 = RandomForestClassifier()
```

```python
model1.fit(x_train, y_train)
```

```python
x_train_pred = model1.predict(x_train)
training_accu = accuracy_score(x_train_pred, y_train)
print('Accuracy on Training data: ', training_accu)
```

```python
x_test_pre = model1.predict(x_test)
testing_accu = accuracy_score(x_test_pre,y_test)
print('Accuracy on Testing data: ', testing_accu)
```
